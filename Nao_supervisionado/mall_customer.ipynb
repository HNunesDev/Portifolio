```python
#Testando a quantidade de cluster melhor
for k in range(2, 11):
    modelo = KMeans(n_clusters=k, random_state=42)
    modelo.fit(df)
    score = silhouette_score(df, modelo.labels_)
    print(k, score)
```

```python
modelo = KMeans(n_clusters=5, random_state=13)
modelo.fit(df)
df['Cluster'] = modelo.labels_
```

```python
df.head()
```

```python
score = silhouette_score(df, modelo.labels_)
print("Silhouette Score:", score)
```

```python
df
```

```python
df1 = df[df.cluster==0] 
df2 = df[df.cluster==1] 
df3 = df[df.cluster==2]
df4 = df[df.cluster==3]
df5 = df[df.cluster==4]

plt.scatter(df1['Annual Income (k$)'],df1['Spending Score (1-100)'],color='green')
plt.scatter(df2['Annual Income (k$)'],df2['Spending Score (1-100)'],color='red')
plt.scatter(df3['Annual Income (k$)'],df3['Spending Score (1-100)'],color='black')
plt.scatter(df4['Annual Income (k$)'],df4['Spending Score (1-100)'],color='c')
plt.scatter(df5['Annual Income (k$)'],df5['Spending Score (1-100)'],color='blue')

plt.scatter(modelo.cluster_centers_[:,0],modelo.cluster_centers_[:,1],
            color='purple',marker='*',label='Centro')

plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
```
